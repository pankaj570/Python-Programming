```python
#Creating Bar Charts
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

#plt.bar(x,y)  # vertical bar
plt.barh(x, y) #Horizontal Bar
plt.show()
```

```python
#Histogram Bar
# Parameters:
# loc: Mean of the distribution (default is 0).
# scale: Standard deviation of the distribution (default is 1).
# size: Output shape or number of samples to be drawn (default is a single value).
```

```python
x = np.random.normal(loc=170,  scale=10, size=250)
plt.hist(x)
plt.show()
```

```python
#Pie Charts
```

```python
y  = np.array([35, 25, 25, 15])
charts_labels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels=charts_labels, autopct='%1.1f%%')
plt.legend(title = "Four Fruits:")
plt.show()
```
